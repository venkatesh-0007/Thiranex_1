```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/loan_approval_dataset.csv')
df.head()
```

```python
df.columns = df.columns.str.strip()
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='loan_status', data=df)
plt.title('Loan Approval Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['income_annum'], bins=30)
plt.title('Income Distribution')
plt.xlabel('Annual Income')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['cibil_score'], bins=30, kde=True)
plt.title('CIBIL Score Distribution')
plt.xlabel('CIBIL Score')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='loan_status', y='loan_amount', data=df)
plt.title('Loan Amount by Loan Status')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='income_annum', y='loan_amount', hue='loan_status', data=df)
plt.title('Income vs Loan Amount')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='education', hue='loan_status', data=df)
plt.title('Education vs Loan Status')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='self_employed', hue='loan_status', data=df)
plt.title('Self Employed vs Loan Status')
plt.show()
```

```python
encoded_df = df.copy()

for col in encoded_df.select_dtypes(include='object').columns:
    encoded_df[col] = encoded_df[col].astype('category').cat.codes

encoded_df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_features = encoded_df.corr()['loan_status'].sort_values(ascending=False)
top_features
```
